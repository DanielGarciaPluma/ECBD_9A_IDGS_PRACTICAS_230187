## 22. Calculando los datos para graficación del total de estadísticas por cada tipo de Pokémon, respetano el orden alfabético previamente definido.

Se aplica un estilo de gráficos accesible y se agrupan los datos de Pokémon por tipo para sumar sus \
estadísticas principales. Luego se generan varios gráficos de barras en una cuadrícula,\
donde se comparan las estadísticas (HP, Attack, Defense, etc.) entre tipos usando diferentes paletas de colores.

````python
plt.style.use('tableau-colorblind10') 

stats=['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
palettes = ['magma','ocean','vlag','copper','mako','winter']
pokemon_stats = pokemons.groupby('Type')[stats].sum()

plt.figure(figsize=(17,17))

for k, stat in enumerate(stats, start=1):
    plt.subplot(3,2,k)
    
    sns.barplot(
        x=pokemon_stats[stat],
        y=pokemon_stats.index,
        hue=pokemon_stats.index, 
        palette=palettes[k-1],
        legend=False
    )
    plt.title(f'Total de {stat}')
    plt.xlabel(stat)
    plt.ylabel('Type')

plt.show()
````

## 23. Visualización con Swarmplots
Se crea una figura que contiene varios gráficos de dispersión, uno para cada estadística de los Pokémon. Mediante un ciclo, se comparan los valores de cada estadística según el tipo de Pokémon, utilizando diferentes colores y una ligera separación entre los puntos para evitar que se sobrepongan. Finalmente, se ajusta la distribución de los gráficos para mejorar su visualización y se muestran en pantalla.


````python
plt.figure(figsize=(15,30))

for k, stat in enumerate(stats, start=1):
    plt.subplot(7,1,k)

    sns.stripplot(
        data=pokemons,
        x='Type',
        y=stat,
        hue='Type',
        palette='Dark2',
        legend=False,
        jitter=True
    )

    plt.title(f'Total de {stat} para cada tipo')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

## 24. Plots para cada tipo de Pokemon

Este código genera una cuadrícula de gráficos de barras, donde cada gráfico representa un tipo de Pokémon. En cada uno se suman las estadísticas de todos los Pokémon de ese tipo y se muestran en un gráfico horizontal con una paleta de colores. Además, se establece un mismo límite en el eje X para facilitar la comparación entre los diferentes tipos de Pokémon.


````python
k=1;
plt.figure(figsize=(17,25))
for i in list_types:
    plt.subplot(7,3,k);   
    k=k+1; 
    type_data = pokemons[pokemons.Type == i].sum(numeric_only=True)
    pokemon_stats = type_data.drop(['#', 'Generation', 'Legendary', 'Total'], errors='ignore')
    
    sns.barplot(
        x=pokemon_stats.values,
        y=pokemon_stats.index,
        hue=pokemon_stats.index,
        palette='inferno',
        legend=False
    )
    
    plt.title(i)
    plt.xlim(0,8500)
````

## 25. Preparación de Datos para graficado de Pokemons

Se utiliza el método .melt() para "derretir" las columnas especificadas en rows.

````python
pok_melt=pd.melt(pokemons,id_vars=['Name','Type','Legendary'],value_vars=['HP','Defense','Attack','Sp. Atk','Sp. Def','Speed','Evolution_Level'])
pok_melt.head()
````

## 26. Graficado con Swarmplots

Genera una cuadrícula de gráficos de dispersión tipo swarmplot, donde cada gráfico corresponde a un tipo de Pokémon. En ellos se muestran los valores de las distintas estadísticas de cada Pokémon, utilizando colores para diferenciarlas. Además, se añade una ligera variación a los datos para evitar la superposición de puntos y mejorar la visualización, ocultando también las advertencias innecesarias de la librería Seaborn.

````python
import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='seaborn')

plt.figure(figsize=(17,25))
k=1

for i in list_types:
    plt.subplot(10, 2, k)
    k = k + 1

    type_melted = pok_melt[pok_melt.Type == i].copy()
    
    if type_melted.empty:
        continue
    
    np.random.seed(42)
    type_melted['value'] = type_melted['value'] + np.random.uniform(-0.1, 0.1, size=len(type_melted))
    
    sns.swarmplot(
        data=type_melted, 
        x='variable', 
        y='value', 
        hue='variable',       
        palette='gist_stern',
        legend=False,          
        size=1.5,             
        alpha=0.7             
    )

plt.tight_layout()
plt.show()
````

## 27. Promedio de la estadísticas agrupadas por el tipo de Pokemon.

Se instancia un Dataframe y se le asignan los datos correspondientes a el promedio de todas las estadísticas, acorde al tipo de Pokemon (agrupadas en base al tipo)

````python
df = pd.DataFrame()
for i in stats:
	df[i]=pokemons.groupby('Type').describe()[i]['mean']

df
````

## 28. Graficado de promedio de cada estadística por cada tipo

Este código crea una serie de gráficos de barras que muestran el **promedio de cada estadística** de los Pokémon por tipo. Mediante un ciclo, genera un gráfico para cada atributo, utilizando una paleta de colores diferente y agregando títulos y etiquetas para facilitar la comparación de los valores promedio entre los distintos tipos de Pokémon.


````python
plt.figure(figsize=(16,35))
k=1
m=0
for i in stats:
    plt.subplot(4,2,k)
    k=k+1
    sns.barplot(x=df[i],hue=df.index,palette=palettes[m])
    m=m+1
    plt.title(str('Promedio del total de ')+ i +str(' para cada tipo'))
    plt.xlabel(i)
````

## 29. Graficado de tipos de pokemones y sus estadísticas

Este código genera una cuadrícula de gráficos de barras donde se muestra el **promedio de las estadísticas** para cada tipo de Pokémon. Cada gráfico utiliza una misma escala y una paleta de colores para facilitar la comparación entre tipos.


````python
k=1;
plt.figure(figsize=(16,25))
for i in list_types:
    plt.subplot(7,3,k);
    k=k+1;
    sns.barplot(x=df.loc[i,:].values,hue=df.loc[i,:].index, palette='Paired');
    plt.title(i)
    plt.xlim(0,130)
    plt.ylabel('Mean')
plt.show()
````

## 30. Ordenamiento descendiente

Este código genera una **gráfica de barras** que muestra la suma de las estadísticas totales de los Pokémon por tipo, ordenadas de mayor a menor y con un título descriptivo.


````python
plt.figure(figsize=(15,5))
sns.barplot(x=pokemons.groupby('Type').sum().Total.sort_values(ascending=False).index
            ,y=pokemons.groupby('Type').sum().Total.sort_values(ascending=False),
            hue=pokemons.groupby('Type').sum().Total.sort_values(ascending=False).index 
            ,legend=False, palette='cool')
plt.title('Total de todas las estadísticas para cada tipo de Pokemon')
````

## 31. Media de todas las estadísticas para cada tipo de Pokemon

Se grafica con barplot la media de todas las estadísticas, para cada tipo de Pokemon en el Dataset. La gráfica se muestra en orden descendente

````python
plt.figure(figsize=(15,5))
sns.barplot(x=pokemons.groupby('Type')["Total"].mean().sort_values(ascending=False).index,
            y=pokemons.groupby('Type')["Total"].mean().sort_values(ascending=False).values,
            hue=pokemons.groupby('Type')["Total"].mean().sort_values(ascending=False).index,
            legend=False,palette='twilight_shifted')
plt.title('Promedio del total de todas las estadísticas para cada tipo de Pokemon')
````

## 40. Plotting (Graficado) con Bokeh

````python
from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
from bokeh.layouts import row, gridplot
output_notebook()

p1 = figure(width=400, height=200, title='HP for each generation')
p1.scatter(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,size=3,color='red')
p1.line(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,line_width=1,color='red')

# Attack
p2 = figure(width=400, height=200, title='HP for each generation')
p2.scatter(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,size=3,color='red')
p2.line(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,line_width=1,color='red')

# Defense
p3 = figure(width=400, height=200, title='HP for each generation')
p3.scatter(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,size=3,color='red')
p3.line(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,line_width=1,color='red')

# Sp. Atk
p4 = figure(width=400, height=200, title='HP for each generation')
p4.scatter(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,size=3,color='red')
p4.line(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,line_width=1,color='red')

# Sp. Def
p5 = figure(width=400, height=200, title='HP for each generation')
p5.scatter(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,size=3,color='red')
p5.line(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,line_width=1,color='red')

# Speed
p6 = figure(width=400, height=200, title='HP for each generation')
p6.scatter(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,size=3,color='red')
p6.line(x=[1,2,3,4,5,6,7],y=pokemons.groupby('Generation').sum().HP,line_width=1,color='red')

grid = gridplot([p1,p2,p3,p4,p5,p6],ncols=2)
show(grid)
````
